```python
from pathlib import Path
import pandas as pd

ls_train_loss = []
ls_test_loss = []

for pth in Path.cwd().glob("Epoch *"):
    ls_train_loss.append((int(pth.name.replace("Epoch ", "")), pd.read_csv(pth / "train-records.csv")["loss"].mean()))
    ls_test_loss.append((int(pth.name.replace("Epoch ", "")), pd.read_csv(pth / "test-records.csv")["loss"].mean()))
```

```python
import matplotlib.pyplot as plt

plt.plot(*zip(*sorted(ls_train_loss)), label = "train")
plt.plot(*zip(*sorted(ls_test_loss)), label = "test")
plt.legend()
```
